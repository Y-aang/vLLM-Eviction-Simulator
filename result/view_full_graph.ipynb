```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_results(file_path):
    """
    
    :param file_path:  ( "results_alpha_1.2.txt")
    """
    try:
        #  ARCCache 
        data = pd.read_csv(file_path, names=["cache_size_fraction", "LRU_hit_rate", "TwoQ_hit_rate", "ARC_hit_rate"])

        # 
        data["cache_size_fraction"] = data["cache_size_fraction"].astype(float)
        data["LRU_hit_rate"] = data["LRU_hit_rate"].astype(float)
        data["TwoQ_hit_rate"] = data["TwoQ_hit_rate"].astype(float)
        data["ARC_hit_rate"] = data["ARC_hit_rate"].astype(float)

        # 
        plt.figure(figsize=(6, 4))
        plt.plot(data["cache_size_fraction"], data["LRU_hit_rate"], marker='o', label="LRUCache Hit Rate", linestyle='-')
        plt.plot(data["cache_size_fraction"], data["TwoQ_hit_rate"], marker='s', label="TwoQCache Hit Rate", linestyle='-')
        plt.plot(data["cache_size_fraction"], data["ARC_hit_rate"], marker='^', label="ARCCache Hit Rate", linestyle='-')

        # 
        for i in range(len(data)):
            x = data["cache_size_fraction"].iloc[i]
            y1 = data["LRU_hit_rate"].iloc[i]
            y2 = data["TwoQ_hit_rate"].iloc[i]
            y3 = data["ARC_hit_rate"].iloc[i]
            plt.text(x, y1 + 0.02, f"{y1:.2f}", ha='center', fontsize=9)
            plt.text(x, y2 + 0.02, f"{y2:.2f}", ha='center', fontsize=9)
            plt.text(x, y3 + 0.02, f"{y3:.2f}", ha='center', fontsize=9)

        # 
        plt.xlabel("Cache Size Fraction (Average Input Length / Cache Size)")
        plt.ylabel("Hit Rate")
        plt.title(f"Cache Performance Comparison\n{file_path}")
        plt.legend()
        plt.grid(True)
        plt.ylim(0, 1)
        
        # 
        plt.show()
    
    except FileNotFoundError:
        print(f"Error: File {file_path} not found. ")
        
plot_results("full_results_alpha_1.0.txt")
```
