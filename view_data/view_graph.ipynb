```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
csv_file = "/Users/shenyang/Desktop/MS Research/workplace/hit_rate_test/view_data/Qwen2.5-1.5B-Instruct_wikiQA_hotspot.csv"  # 请将文件路径替换为你的CSV文件路径

# 读取 CSV 数据
df = pd.read_csv(csv_file)

# 绘制 Hitrate 图
plt.figure(figsize=(10, 6))
plt.plot(df['Cache_Size'], df['ARC_hitrate'], label='ARC', marker='o')
plt.plot(df['Cache_Size'], df['DBL_hitrate'], label='DBL', marker='o')
plt.plot(df['Cache_Size'], df['LRU_hitrate'], label='LRU', marker='o')
plt.title('Hit Rate on different C_size / Avg_prompt_length')
plt.xlabel('C_size / Avg_prompt_length')
plt.ylabel('Hit Rate')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 绘制 Time 图
plt.figure(figsize=(10, 6))
plt.plot(df['Cache_Size'], df['ARC_time'], label='ARC', marker='o')
plt.plot(df['Cache_Size'], df['DBL_time'], label='DBL', marker='o')
plt.plot(df['Cache_Size'], df['LRU_time'], label='LRU', marker='o')
plt.title('Avg_TTFT on different C_size / Avg_prompt_length')
plt.xlabel('Cache Size')
plt.ylabel('TTFT (s)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
