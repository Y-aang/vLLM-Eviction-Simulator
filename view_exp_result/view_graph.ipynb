```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_cache_results(csv_file, hitrate_title="Hit Rate on different C_size / Avg_prompt_length", 
                       time_title="Avg_TTFT on different C_size / Avg_prompt_length"):
    """
    读取 CSV 文件并绘制 Hit Rate 和 TTFT 图。
    
    参数:
        csv_file (str): CSV 文件路径
        hitrate_title (str): Hit Rate 图标题
        time_title (str): TTFT 图标题
    """
    df = pd.read_csv(csv_file)

    # 绘制 Hitrate 图
    plt.figure(figsize=(10, 6))
    plt.plot(df['Cache_Size'], df['ARC_hitrate'], label='ARC', marker='o')
    plt.plot(df['Cache_Size'], df['DBL_hitrate'], label='DBL', marker='o')
    plt.plot(df['Cache_Size'], df['LRU_hitrate'], label='LRU', marker='o')
    plt.title(hitrate_title)
    plt.xlabel('C_size / Avg_prompt_length')
    plt.ylabel('Hit Rate')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # 绘制 Time 图
    plt.figure(figsize=(10, 6))
    plt.plot(df['Cache_Size'], df['ARC_time'], label='ARC', marker='o')
    plt.plot(df['Cache_Size'], df['DBL_time'], label='DBL', marker='o')
    plt.plot(df['Cache_Size'], df['LRU_time'], label='LRU', marker='o')
    plt.title(time_title)
    plt.xlabel('C_size / Avg_prompt_length')
    plt.ylabel('TTFT (s)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# WikiQA
plot_cache_results(
    "/Users/shenyang/Desktop/MS Research/workplace/hit_rate_test/view_data/Qwen2.5-1.5B-Instruct_wikiQA_hotspot.csv",
    hitrate_title="[WikiQA 142 Docs ~12k Tokens] Hit Rate on Cache Ratio (150 Requests, Alpha=1, HotSpot: 30 Window Size, 10% ×10 Boost)",
    time_title="[WikiQA 142 Docs ~12k Tokens] Avg TTFT on Cache Ratio (150 Requests, Alpha=1, HotSpot: 30 Window Size, 10% ×10 Boost)"
)
```
